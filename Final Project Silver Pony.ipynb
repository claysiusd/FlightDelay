# Undersampling

```python
from sklearn.utils import resample

X = new_data[['DAY_OF_MONTH', 'DAY_OF_WEEK', 'OP_CARRIER_AIRLINE_ID', 
            'OP_CARRIER_FL_NUM', 'ORIGIN_AIRPORT_ID', 'DEST_AIRPORT_ID', 
            'YEAR', 'Distance_Norm', 'DEP_TIME_BLK_LABEL', 'TAIL_CODE_LABEL']]

y = new_data[['DEP_DEL15']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 28)
```

```python
X = pd.concat([X_train, y_train], axis=1)

non_delay = X[X.DEP_DEL15 == 0]
delay = X[X.DEP_DEL15 == 1]
```

```python
non_delay_under = resample(non_delay, replace = False,
                     n_samples = len(delay),
                     random_state = 28)

data_under = pd.concat([non_delay_under, delay])

data_under.DEP_DEL15.value_counts()
```

```python
X_train = data_under.drop('DEP_DEL15', axis=1)
y_train = data_under.DEP_DEL15
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(random_state=28)
logreg.fit(X_train, y_train)

y_predicted = logreg.predict(X_test)

from sklearn.metrics import classification_report
print('\nclassification report')
print(classification_report(y_test, y_predicted))

from sklearn.metrics import roc_curve, auc
print('\nAUC report')
fpr, tpr, thresholds = roc_curve(y_test, y_predicted, pos_label=1)
print(auc(fpr, tpr))
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier

#train
dt = DecisionTreeClassifier(random_state=28)
dt.fit(X_train,y_train)

#predict
y_predicted = dt.predict(X_test)

#eval
from sklearn.metrics import classification_report
print('\nclassification report')
print(classification_report(y_test, y_predicted)) 

from sklearn.metrics import roc_curve, auc
print('\nAUC report')
fpr, tpr, thresholds = roc_curve(y_test, y_predicted, pos_label=1)
print(auc(fpr, tpr))
```
